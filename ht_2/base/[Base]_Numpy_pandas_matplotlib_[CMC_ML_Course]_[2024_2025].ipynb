```python
df.publishedAt = pd.to_datetime(df.publishedAt)
```

```python
df.publishedAt.dt.date
```

```python
unique_videos = df.drop_duplicates(subset='publishedAt')

unique_videos.loc[:, 'publishedAt'] = pd.to_datetime(unique_videos.publishedAt).dt.date

videos = unique_videos.groupby('publishedAt').size().reset_index(name='videos_count')

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

plt.plot(videos.publishedAt, videos.videos_count, marker='o', color='red')

plt.title('     ', fontsize= 14)
plt.xlabel(' ')
plt.ylabel('  ')

plt.xticks(ticks=videos.publishedAt[::3],rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
videos_in_countrys = df.groupby('country').size().reset_index(name='video_count')
videos_in_countrys = videos_in_countrys.sort_values(by='video_count', ascending=False)

plt.figure(figsize=(12, 8))
plt.bar(videos_in_countrys.country, videos_in_countrys.video_count, color='mediumseagreen')

plt.title('      ', fontsize=14)
plt.xlabel('')
plt.ylabel(' ')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

```python
france = df[df['country'] == 'FR']
france = france[france['comments_disabled'] == False]


plt.figure(figsize=(12,8))
plt.scatter(france.view_count, france.comment_count, alpha=0.8)

plt.title(' -   -     ', fontsize=14)
plt.xlabel('  ( )', fontsize=12)
plt.ylabel('  ( )', fontsize=12)

plt.xscale('log')
plt.yscale('log')

plt.tight_layout()
plt.show()
```
